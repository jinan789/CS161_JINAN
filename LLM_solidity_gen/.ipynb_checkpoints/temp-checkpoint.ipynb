```python
def write_to_file(contents, path):
    with open(path, 'w') as f:
        f.write(contents)
    
def read_from_file(path):
    with open(path) as f:
        return ''.join(f.readlines())
    
def find_sub_list(sl,l):
    sll=len(sl)
    for ind in (i for i,e in enumerate(l) if e==sl[0]):
        if l[ind:ind+sll]==sl:
            return ind,ind+sll-1
    return -9999, -9999
```

```python
# Find evolvements across generations
# def find_diff_across_gens

modified_count = 0
no_modified_count = 0
for func_name in func_name_key_set:
    
    def stripped(lst):
        return [i.strip() for i in lst]
    
    if func_name not in list_of_long_enough_functions:
        continue
        
    print(func_name)
    
    target_generation_path = 'prompts_folder/new_function_dataset_builder/deduplicated_by_func_name/' + 'DIRECT_GENERATED_' + func_name + '.txt'
    direct_generated = read_from_file(target_generation_path)
    
    
    cur_code_path = 'prompts_folder/new_function_dataset_builder/deduplicated_by_func_name/' + 'CODES_AFTER_COT_' + func_name + '.txt'
    codes_after_cot = read_from_file(cur_code_path)

    codes_before = codes_after_cot.split('\n')
    codes_after = direct_generated.split('\n')
    
    after_diff = [i for i in codes_after if i.strip() not in stripped(codes_after)]
    before_diff = [i for i in codes_after if i.strip() not in stripped(codes_before)]
    
    after_diff = [i for i in after_diff if not i.strip().startswith('//')]
    before_diff = [i for i in before_diff if not i.strip().startswith('//')]
    
    if len(after_diff) == len(before_diff) and len(before_diff) == 0:
        no_modified_count += 1
        continue
    else:
        modified_count += 1
        print(after_diff)
        print(before_diff)
        print('\n\n')
    
assert no_modified_count + modified_count == 53
    #break
```
