```python
import random
```

```python
def get_3_dice():
    return [random.randint(1,6) for _ in range(3)]

def get_3_dice_with_result():
   
    result = dict()     
    t = [random.randint(1,6) for _ in range(3)]
    s = sum(t)
    result['dices'] = t
    result['sum'] = s
    result['is_large'] = s >= 11 and s <= 17
    result['is_small'] = s >= 4 and s <= 10
    result['is_neutral'] = s == 3 or s == 18
    return result

def play_N_games(N, total_cash, strategy, cash_records):
    result = {}
    losses = 0
    is_win = False
    bid = 0
    
    for game_id in range(N):
        result[game_id] = dict()
        cash_records.append(total_cash)
        
        bid, losses = strategy(losses, is_win, bid)
        if bid > total_cash:
            return result, total_cash, cash_records
        
        t = get_3_dice_with_result()
        result[game_id]['table_info'] = t
        is_win = t['is_small']
        if is_win:
            losses = 0
            total_cash += bid
        else:
            losses += bid
            total_cash -= bid
        result[game_id]['player_info'] = (losses, is_win, bid)
        
        print('At game ', game_id,  ';   bid: ', bid, ";  result is :", is_win, ';   cash left: ', total_cash)
       
    cash_records.append(total_cash)
    return result, total_cash, cash_records

def strategy_safe(losses_before_this, previous_is_win, previous_bid):
    # returns the amount to bid
    # previous_is_win: Last bid was win?
    
    # analysis
    # 100, 100, 200, 400, 800 (total is 1600)
    # stop at 800, probability distribution is:
    #    Pr[win 100] = 1/2
    #.   Pr[lose 1600] = 1/32
    #.   Pr[no win no lose] = 15/32
    #. 
    #.   E[return] = 50 - 50 = 0
    #
    if previous_bid == 0:
        # first bid
        return 100, 0
    
    if previous_is_win:
        return 100, 0
        
    if previous_bid == 200:
        return 100, 0
    else:
        return losses_before_this, losses_before_this
        
    
```

```python
cash = 2000
cash_records = []

_, cash, cash_records = play_N_games(1000, cash, strategy_safe, cash_records)
```
