```python
import os
```

```python
def read_from_file(path):
    with open(path) as f:
        return ''.join(f.readlines())
    
def write_to_file(contents, path):
    with open(path, 'w') as f:
        f.write(contents)
```

```python
gas_cost_of_patterns = [l for l in read_from_file('tables_count/gas_cost_of_patterns.txt').split('\n') if l != '']
gas_cost_of_patterns = [i for i in gas_cost_of_patterns if i.startswith('GAS') or i.startswith('STEP')]
gas_cost_of_patterns = [i for i in gas_cost_of_patterns if ' , , , ,' not in i]
```

```python
new_gas_cost_of_patterns = []
for l in gas_cost_of_patterns:
    new_gas_cost_of_patterns.append([i.strip() for i in l.split(':')[1].strip().split(',') if i.strip() != ''])
```

```python
new_gas_cost_of_patterns
```

```python
gas_cost_of_patterns
```

```python
deployment_pattern_num_to_gas_step = dict()
running_pattern_num_to_gas_step = dict()

for i in range(1,len(gas_cost_of_patterns)//2+1):
    gas_line_ind = (i - 1) * 2
    step_line_ind = (i - 1) * 2 + 1
    
    d_dep = dict()
    d_run = dict()
    deployment_pattern_num_to_gas_step[i] = d_dep
    running_pattern_num_to_gas_step[i] = d_run
    
    d_dep['gas_before'] = new_gas_cost_of_patterns[gas_line_ind][0]
    d_dep['gas_after'] = new_gas_cost_of_patterns[gas_line_ind][1]
    d_dep['step_before'] = new_gas_cost_of_patterns[step_line_ind][0]
    d_dep['step_after'] = new_gas_cost_of_patterns[step_line_ind][1]
    d_dep['step_saved'] = -int(d_dep['step_after']) + int(d_dep['step_before'])
    d_dep['gas_saved'] = -int(d_dep['gas_after']) + int(d_dep['gas_before'])
    d_dep['gas_saved_percentage'] = (-int(d_dep['gas_after']) + int(d_dep['gas_before'])) / int(d_dep['gas_before'])
    
    d_run['gas_before'] = new_gas_cost_of_patterns[gas_line_ind][2]
    d_run['gas_after'] = new_gas_cost_of_patterns[gas_line_ind][3]
    d_run['step_before'] = new_gas_cost_of_patterns[step_line_ind][2]
    d_run['step_after'] = new_gas_cost_of_patterns[step_line_ind][3]
    d_run['step_saved'] = -int(d_run['step_after']) + int(d_run['step_before'])
    d_run['gas_saved'] = -int(d_run['gas_after']) + int(d_run['gas_before'])
    d_run['gas_saved_percentage'] = (-int(d_run['gas_after']) + int(d_run['gas_before'])) / int(d_run['gas_before'])

    
```

```python
cur_file_path = 'tables_count/' + 'actual_run.txt'
full_table_text = read_from_file(cur_file_path)
```

```python
data_entry_rows = [i for i in full_table_text.split('\n') if i.strip().endswith('\\') and '&' in i]
pattern_name_to_RSM = dict()

for r in data_entry_rows:
    items = r.split('&')
    print(items)
    assert len(items) == 10 or len(items) == 5 or len(items) == 8
    if '\\textbf{Pattern name}   ' in items:
        continue
    if items[0].strip() == '':
        continue
    
    pattern_name = items[0].strip()
    round_num = items[1].strip().split('.')[0]
    contract_num = items[1].strip().split('.')[1]
    verification_category = items[2].strip()
    passed = items[3].strip()
    
    if verification_category.startswith('A'):
        assert passed == 'N'
    else:
        assert passed == 'Y'
        

    R = items[4].strip()
    S = items[5].strip()
    M = items[6].strip()

    temp_l = pattern_name_to_RSM.get(verification_category, [])
    temp_l.append((R,S,M))
    pattern_name_to_RSM[verification_category] = temp_l
    
    
```

```python
pattern_name_to_RSM.keys()
```

```python
for i in pattern_name_to_RSM:
    R = []
    S = []
    M = []
    for T in pattern_name_to_RSM[i]:
        if T[0] == 'NA':
            assert T[1] == 'NA' and T[2] == 'NA'
            continue
        R.append(float(T[0]))
        S.append(float(T[1]))
        M.append(float(T[2]))
    if len(R) == 0:
        assert(len(S) == 0 and len(M) == 0)
        continue
    R = sum(R)/len(R)
    S = sum(S)/len(S)
    M = sum(M)/len(M)
    
    pattern_name_to_RSM[i] = (R,S,M)
```

```python
to_print = ''
for i in running_pattern_num_to_gas_step:
    to_print += 'Pattern ' + str(i) + '.  '
    to_print += ' & ' + deployment_pattern_num_to_gas_step[i]['gas_before']
    to_print += ' & ' + deployment_pattern_num_to_gas_step[i]['gas_after']
    to_print += ' & ' + str(deployment_pattern_num_to_gas_step[i]['gas_saved_percentage']*100)[:6] + ' \\%'
    
    to_print += ' & ' + running_pattern_num_to_gas_step[i]['gas_before']
    to_print += ' & ' + running_pattern_num_to_gas_step[i]['gas_after']
    to_print += ' & ' + str(running_pattern_num_to_gas_step[i]['gas_saved_percentage']*100)[:6] + ' \\%'
    
    to_print += '\\\\ \n \\hline \n\n'
    
```
