```python
def read_from_file(path):
    with open(path) as f:
        return ''.join(f.readlines())
```

```python
data_entry_rows = [i for i in full_table_text.split('\n') if i.strip().endswith('\\') and '&' in i]
round_to_item_dict = dict()

for r in data_entry_rows:
    items = r.split('&')
    print(items)
    assert len(items) == 10 or len(items) == 5
    if '\\textbf{Pattern name}   ' in items:
        continue
    
    
    pattern_name = items[0].strip()
    round_num = items[1].strip().split('.')[0]
    contract_num = items[1].strip().split('.')[1]
    verification_category = items[2].strip()
    passed = items[3].strip()
    
    if verification_category.startswith('A'):
        assert passed == 'N'
#     R = items[].strip()
#     S = items[].strip()
#     M = items[].strip()
#     P_categories = items[].strip()
#     S_categorise = items[].strip()
#     comments = items[].strip()

#     if int(round_num) in [1,2,3]:
#         round_num = 1
#     elif int(round_num) in [4,5,6]:
#         round_num = 2
#     elif int(round_num) in [7,8]:
#         round_num = 3
#     else:
#         raise Exception()
    cur_l = round_to_item_dict.get(round_num, [])
    cur_l.append((pattern_name, contract_num, verification_category, passed))
    round_to_item_dict[round_num] = cur_l
    
    
```

```python
import math
```

```python
round_to_stats_dict = dict()
max_dict = dict()
cols = ['pass_rate', 'fail_rate', 'num_pattern_categories', 'num_failure_categories', 'num_reported_patterns', 
        'num_reported_patterns_passed', 'num_reported_patterns_failed', 'pattern_entropy', 'failure_entropy']


for round_num in round_to_item_dict:
    cur_d = dict()
    num_passed = 0
    num_failed = 0
    pattern_categories = []
    failure_categories = []
    for item in round_to_item_dict[round_num]:
        pattern_name, contract_num, verification_category, passed = item
        if passed == 'Y':
            num_passed += 1
        elif passed == 'N':
            num_failed += 1
        else:
            print(item)
            raise Exception()
        if verification_category.startswith('A'):
            failure_categories.append(verification_category)
        else:
            pattern_categories.append(verification_category)
    cur_d['pass_rate'] = num_passed / (num_passed + num_failed)
    cur_d['fail_rate'] = num_failed / (num_passed + num_failed)
    cur_d['num_pattern_categories'] = len(set(pattern_categories))
    cur_d['num_failure_categories'] = len(set(failure_categories))
    cur_d['num_reported_patterns'] = len(pattern_categories) + len(failure_categories)
    cur_d['num_reported_patterns_passed'] = len(pattern_categories)
    cur_d['num_reported_patterns_failed'] = len(failure_categories)
    
    
    entropy_dict = dict()
    for p in pattern_categories:
        cur_c = entropy_dict.get(p, 0)
        entropy_dict[p] = cur_c + 1
    pattern_entropy = 0
    total_length_to_divide_by = len(pattern_categories)
    for p in entropy_dict:
        p_i = entropy_dict[p] / total_length_to_divide_by
        pattern_entropy -= p_i * math.log2(p_i)
    
    cur_d['pattern_entropy'] = pattern_entropy
    
    
    entropy_dict = dict()
    for p in failure_categories:
        cur_c = entropy_dict.get(p, 0)
        entropy_dict[p] = cur_c + 1
    failure_entropy = 0
    total_length_to_divide_by = len(failure_categories)
    for p in entropy_dict:
        p_i = entropy_dict[p] / total_length_to_divide_by
        failure_entropy -= p_i * math.log2(p_i)
    
    cur_d['failure_entropy'] = failure_entropy
        
    
    
    round_to_stats_dict[round_num] = cur_d
    
    
    
    for c in cols:
        if c not in max_dict:
            max_dict[c] = cur_d[c]
        else:
            max_dict[c] = max([cur_d[c], max_dict[c]])
            
t_l = []
for cur_num in range(1,9):
    l = []
    for f in os.listdir('api_broken/questions_repo_2/round_' + str(cur_num)):
        f = 'api_broken/questions_repo_2/round_' + str(cur_num) + '/' + f
        l.append(len(read_from_file(f).split('\n')))
        
    round_to_stats_dict[str(cur_num)]['avg_num_lines'] = sum(l) / len(l)
    t_l.append(sum(l) / len(l))
max_dict['avg_num_lines'] = max(t_l)
    

```

```python
contents = ''
cols = ['pass_rate', 'fail_rate', 'num_pattern_categories', 'num_failure_categories', 'num_reported_patterns', 
        'num_reported_patterns_passed', 'num_reported_patterns_failed', 'pattern_entropy', 'failure_entropy', 
        'avg_num_lines']

for i in round_to_stats_dict:
    cur_row = str(i) + ' ' * 3 + '&'
    for c in cols:
        s = round_to_stats_dict[i][c]        
        if type(s) == float:
            new_s = round(s, 3)
        else:
            new_s = s
        if s == max_dict[c]:
            new_s = '\\textbf{' + str(new_s) + '}'
            
        cur_row += ' ' * 3  + str(new_s) + ' ' * 3
        if c == cols[-1]:
            cur_row += '\\\\' + '\n'
        else:
            cur_row += '&'
    contents += cur_row
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
font1 = {
    'family' : 'Times New Roman',
#     'family' : 'Arial',
#     'family' : 'SimHei',
    'weight' : 'normal',
    'size'   : 36,
}
def make_plot_of_dict(d, x_label, y_label, title, save_as):

    d = list(reversed(sorted(list(d.items()), key = lambda x:x[0])))
    xs = [i[0] for i in d]
    ys = [i[1] for i in d]
    
    fig = plt.figure(figsize = (10, 5))

    # creating the bar plot
    plt.bar(xs, ys, color ='maroon',
            width = 0.4)

    #plt.xticks(fontsize=37)  # x轴刻度
    #plt.yticks(fontsize=37)  # y轴刻度

    plt.rcParams['xtick.direction'] = 'in'  #x轴刻度向内
    plt.rcParams['ytick.direction'] = 'in'  #y轴刻度向内
    plt.xticks(fontsize=20)  # x轴刻度
    plt.yticks(fontsize=20)  # y轴刻度
    
    ax=plt.gca()  #获得坐标轴的句柄
    ax.spines['bottom'].set_linewidth(3)  #设置底部坐标轴的粗细
    ax.spines['left'].set_linewidth(3)  #设置左边坐标轴的粗细
    ax.spines['right'].set_linewidth(3)  #设置右边坐标轴的粗细
    ax.spines['top'].set_linewidth(3)  #设置上部坐标轴的粗细
    ax.tick_params(which='both', width=3, pad=10)
    ax.tick_params(which='major', length=9)
    ax.tick_params(which='minor', length=5)
    
    plt.xlabel(x_label, fontsize=37,labelpad=20)
    plt.ylabel(y_label, fontsize=37,labelpad=20)
    plt.title(title)
    plt.show()
    fig.savefig(save_as + ".pdf", bbox_inches='tight')

```

```python
passed_type_to_line = dict()
failed_type_to_line = dict()
for cur_rount in rounds:
    cur_round_passed_type_to_line = dict()
    cur_round_failed_type_to_line = dict()
    cur_round_passed_type_to_count = dict()
    cur_round_failed_type_to_count = dict()
    
    print(cur_rount)
    cur_passes = 0
    cur_failes = 0
    
    cur_file_path = 'tables_count/' + cur_rount
    cur_txt = read_from_file(cur_file_path)
    cur_lines = cur_txt.split('\n')
    cur_lines = [i for i in cur_lines if i.strip().endswith('\\')]
    cur_lines = [[j.strip() for j in i.split('&')] for i in cur_lines]
    cur_lines = [i for i in cur_lines if i[3] == 'N' or i[3] == 'Y']
    for line in cur_lines:
        if line[3] == 'Y':
            cur_passes += 1
            assert 'A' not in line[2]
            l = passed_type_to_line.get(line[2], [])
            l.append(list(line))
            passed_type_to_line[line[2]] = l
            
            l = cur_round_passed_type_to_line.get(line[2], [])
            l.append(list(line))
            cur_round_passed_type_to_line[line[2]] = l
            
            
        else:
            cur_failes += 1
            assert line[2].startswith('A')
            l = failed_type_to_line.get(line[2], [])
            l.append(list(line))
            failed_type_to_line[line[2]] = l
            
            l = cur_round_failed_type_to_line.get(line[2], [])
            l.append(list(line))
            cur_round_failed_type_to_line[line[2]] = l
            

    for i in cur_round_passed_type_to_line:
        cur_round_passed_type_to_count[i] = len(cur_round_passed_type_to_line[i])
    for i in cur_round_failed_type_to_line:
        cur_round_failed_type_to_count[i] = len(cur_round_failed_type_to_line[i])

    print('pass rate: ', cur_passes/(cur_passes + cur_failes))
    print('out of total cases: ', (cur_passes + cur_failes), '\n')
    
    print(cur_round_passed_type_to_count)
    print(cur_round_failed_type_to_count)
    
    make_plot_of_dict(cur_round_passed_type_to_count, 'type_ind', 'counts', cur_rount, cur_rount + '_passed')
    make_plot_of_dict(cur_round_failed_type_to_count, 'type_ind', 'counts', cur_rount, cur_rount + '_failed')
            
passed_type_to_count = dict()
failed_type_to_count = dict()

for i in passed_type_to_line:
    passed_type_to_count[i] = len(passed_type_to_line[i])
for i in failed_type_to_line:
    failed_type_to_count[i] = len(failed_type_to_line[i])
    
make_plot_of_dict(passed_type_to_count, 'type_ind', 'counts', 'overall', 'overall_passed')
make_plot_of_dict(failed_type_to_count, 'type_ind', 'counts', 'overall', 'overall_failed')
```
