```python
def split_code_with_max_m_chars(contract_obj, m, contract_names_mapping_step_4):
    prefix = lines_to_str(contract_obj.get_prefix_for_functions())
    postfix = lines_to_str(contract_obj.get_postfix_for_functions())

    length_of_boundaries = len(prefix) + len(postfix)
    cur_s = ''
    
    t, cur_select_contract_copy = contract_names_mapping_step_4[contract_obj.get_contract_name()]

    target_f_names = [i[0] for i in t]
    found_any = False
    used_func = None
    # first put in one of the functions
    for f in self.function_lst:
        f_name = f.get_sig_code_lines()[0].split('function ')[1].split('(')[0].strip()
        if f_name in target_f_names:
            new_f = lines_to_str(f.get_full_code_lines())
            cur_s += new_f
            found_any = True
            used_func = f
            break
    assert found_any
    assert used_func is not None
    
    
    for f in self.function_lst:
        if f is used_func:
            continue 
            
        if f.is_pure_sig:
            continue

        if len(f.get_body_code_lines()) <= 4:
            continue
        else:
            new_f = lines_to_str(f.get_full_code_lines())
            if len(cur_s) + len(new_f) + length_of_boundaries > m:
                return prefix + cur_s + new_f + postfix
            else:
                cur_s += new_f
    assert len(cur_s) != 0
    return prefix + cur_s + new_f + postfix
```
